```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\Admin\Downloads\2nd project\Luxury_Housing_Bangalore.csv"
)

df.head()
```

```python
##Fix column names##
```

```python
df.columns
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
df.columns
```

```python
##CHECK MISSING VALUES##
```

```python
missing = df.isnull().sum()
missing[missing > 0]
```

```python
df.isnull().sum()
```

```python
##CHECK DATA FORMAT BEFORE CLEANING##
```

```python
df['ticket_price_cr'].unique()[:20]
```

```python
df['ticket_price_cr'].dtype
```

```python
df['ticket_price_cr'] = (
    df['ticket_price_cr']
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace("Cr", "", regex=False)
    .str.strip()
    .astype(float)
)
```

```python
df['ticket_price_cr'].unique()[:20]
```

```python
##HANDLE MISSING VALUES##
```

```python
df['amenity_score'] = df['amenity_score'].fillna(df['amenity_score'].median())
```

```python
df.columns
```

```python
df['transaction_type'].value_counts()
```

```python
df['transaction_type'] = df['transaction_type'].fillna('Unknown')
```

```python
df.transaction_type
```

```python
df['possession_status'] = df['possession_status'].fillna('Under Planning')
```

```python
text_cols = [
    'micro_market',
    'developer_name',
    'configuration',
    'buyer_type',
    'sales_channel'
]

for col in text_cols:
    df[col] = df[col].str.strip().str.title()
```

```python
##FEATURE ENGINEERING## ##KPL'S##
```

```python
##Booking flag##
```

```python
df['booking_flag'] = df['transaction_type'].apply(
    lambda x: 1 if str(x).lower() == 'booked' else 0
)

df[['transaction_type', 'booking_flag']].head(10)
```

```python
##Price per sqft##
```

```python
df['price_per_sqft'] = (
    df['ticket_price_cr'] * 10_000_000 / df['unit_size_sqft']
)
```
