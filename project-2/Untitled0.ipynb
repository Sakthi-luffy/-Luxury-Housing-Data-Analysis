```python
import pandas as pd

df = pd.read_csv("luxury_housing_cleaned.csv")
df.head()
```

```python
import sqlite3

conn = sqlite3.connect("luxury_housing.db")
df.to_sql("housing_sales", conn, if_exists="replace", index=False)

print("Table created successfully!")
```

```python
query = "SELECT COUNT(*) AS total_rows FROM housing_sales"
pd.read_sql(query, conn)
```

```python
CREATE DATABASE IF NOT EXISTS luxury_housing;
USE luxury_housing;

CREATE TABLE IF NOT EXISTS housing_sales (
    Property_ID VARCHAR(50),
    Micro_Market VARCHAR(100),
    Project_Name VARCHAR(150),
    Developer_Name VARCHAR(150),
    Unit_Size_Sqft INT,
    Configuration VARCHAR(20),
    Ticket_Price_Cr FLOAT,
    Transaction_Type VARCHAR(50),
    Buyer_Type VARCHAR(50),
    Purchase_Quarter VARCHAR(20),
    Connectivity_Score FLOAT,
    Amenity_Score FLOAT,
    Possession_Status VARCHAR(50),
    Sales_Channel VARCHAR(50),
    NRI_Buyer VARCHAR(10),
    Locality_Infra_Score FLOAT,
    Avg_Traffic_Time_Min INT,
    Buyer_Comments TEXT,
    Price_per_Sqft FLOAT,
    Quarter_Number INT,
    Booking_Flag INT
);

```
